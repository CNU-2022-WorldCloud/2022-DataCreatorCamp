```python
import os
import shutil
 
original_dataset_dir = '/content/drive/MyDrive/ColabNotebooks/'   
classes_list = os.listdir(original_dataset_dir) 
 
base_dir = '/content/drive/MyDrive/ColabNotebooks/temp/'
os.mkdir(base_dir)
 
train_dir = os.path.join(base_dir, 'train') 
os.mkdir(train_dir)
validation_dir = os.path.join(base_dir, 'val')
os.mkdir(validation_dir)
```

```python
for cls in classes_list:     
    os.mkdir(os.path.join(train_dir, cls))
    os.mkdir(os.path.join(validation_dir, cls))
```

```python
#https://jonsyou.tistory.com/23
#https://velog.io/@jee-9/Scikit-learn-%EA%B5%90%EC%B0%A8-%EA%B2%80%EC%A6%9DCross-validation-KFold-Stratified-KFold-crossvalscore

import math
from sklearn.model_selection import KFold


i = 0

for cls in classes_list:

  train_list = []
  validation_list = []


  path = os.path.join(original_dataset_dir, cls)
  one_class_files = os.listdir(path)

  kf = KFold(n_splits=10, random_state=0, shuffle=True)

  if(len(one_class_files) == 0):
    break;
    
  for train, validation in kf.split(one_class_files):
    
    train_list.append(train)
    validation_list.append(validation)
      

  for temp_arr in train_list:
    print(temp_arr)
    for one_train in temp_arr:
      src = os.path.join(path, one_class_files[one_train])
      dst = os.path.join(os.path.join(train_dir, cls), one_class_files[one_train])
      shutil.copyfile(src, dst)

  for temp_arr in validation_list:
    print(temp_arr)
    for one_validation in temp_arr:
      src = os.path.join(path, one_class_files[one_validation])
      dst = os.path.join(os.path.join(validation_dir, cls), one_class_files[one_validation])
      shutil.copyfile(src, dst)

  i = i+1

  print(path)
```
